```python
# firstly we import the libraries
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
#importing the image
img_path="/content/minepic.jpg"
img=cv2.imread(img_path)
print(img.shape)
```

```python
# now converting the image into grayscale
gray_scale=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_scale,cmap='gray')
```

```python
# croping the image
x = 150  # starting x coordinate
y = 250  # starting y coordinate
width = 300  # width of the ROI
height = 400  # height of the ROI
# Crop the image using the ROI coordinates
cropped_img = img[y:y+height, x:x+width]
# Display the cropped image using matplotlib
plt.imshow(cv2.cvtColor(cropped_img, cv2.COLOR_BGR2RGB))
plt.axis('on')
plt.show()
```

```python
#bluring the image
output2=cv2.blur(gray_scale,(10,10))
plt.imshow(output2,cmap='gray')
plt.show()
```

```python
#rotating the image
angle=30
scale=1.0
(h,w)=img.shape[:2]
center=(w//2,h//2)
M = cv2.getRotationMatrix2D(center, angle, scale)
# Apply the rotation to the image
rotated_img = cv2.warpAffine(gray_scale, M, (w, h))
# Display the rotated image
plt.imshow(cv2.cvtColor(rotated_img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
